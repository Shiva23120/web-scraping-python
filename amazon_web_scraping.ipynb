```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
link='https://www.amazon.in/Top-Examples-Use-Essay-Evidence/product-reviews/1479248738'
page=requests.get(link)
page
```

```python
print(page.content)
```

```python
soup=bs(page.content,'html.parser')
soup
```

```python
print(soup.prettify())
```

```python
name=soup.find_all('span',class_='a-profile-name')
name
```

```python
len(name)
```

```python
cnames=[]
for i in range(len(name)):
    cnames.append(name[i].get_text())
```

```python
cnames
```

```python
titles=soup.find_all('a',class_='review-title-content')
titles
```

```python
title_names=[]
for i in range(len(titles)):
    title_names.append(titles[i].get_text())
```

```python
title_names[:]=[i.lstrip('\n') for i in title_names]
title_names
```

```python
title_names[:]=[i.rstrip('\n') for i in title_names]
print(title_names,end=' ')
```

```python
rating=soup.find_all('i',class_='review-rating')
rate=[]
for i in range(len(rating)):
    rate.append(rating[i].get_text())
    
```

```python
rate
```

```python
reviews=soup.find_all('span',{'data-hook':"review-body"})
reviews
```

```python
rew=[]
for j in range(len(reviews)):
    rew.append(reviews[j].get_text())
    
```

```python
rew
```

```python
soup.title.get_text()
```

```python
cnames,title_names,rate,rew
```

```python
rew[:]=[i.lstrip('\n') for i in rew]
rew[:]=[i.rstrip('\n') for i in rew]
```

```python
import pandas as pd
```

```python
df=pd.DataFrame()
df
```

```python
 df['cnames']=cnames
```

```python
df['rating']=rate
df['review']=rew
```
